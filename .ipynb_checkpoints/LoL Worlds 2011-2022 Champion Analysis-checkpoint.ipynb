Notice the "event" column. I am only interested in the main event (not the play-ins), since the play-ins teams are not considered to be the best of the best in the world at their respective times. As such, I will only be concerning myself with the main event statistics.

```python
champs = champs[champs["event"] == "Main"]
```

Are there any NAs in this dataset?

```python
champs.isna().sum()
```

There may be NAs because LoL was in its infancy in its early seasons. Since it can be argued that seasons 1-4 are LoL's infancy and the beginning of modern LoL starts in season 5 (due to the map changing to the modern one we see today), I will choose to delete those rows.

```python
champs = champs[champs["season"] >= 5]
champs
```

I still notice some NAs. In fact, upon closer inspection, these are all NaNs.

```python
champs.isna().sum()
```

The reason why we still have some NaNs is because the statistic does not exist. For example, take Kassadin from Season 6

```python
champs.loc[(champs["season"] == 6) & (champs["champion"] == "Kassadin")]
```

If we look at, for example, the number of bans, number of games played, and played by number of players, we notice that there were 4 bans, 0 games, and an NaN number of players that played the champion. Obviously, this value of "played_by_number_of_players" should be 0.

```python
champs.loc[(champs["season"] == 6) & (champs["champion"] == "Kassadin")][["season", "champion", "banned_games", "played_games", "played_by_number_of_players"]]
```

```python
champs = champs.replace(np.nan, 0)
champs.isna().sum()
```

### Question: What champion was the most overpowered in each season?

One way we can answer this question is by the pick ban ratio. Any champ with the highest pick ban ratio is the highest contested pick in the tournament. Thus, it should be the most overpowered. Let's first try with Season 5.

```python
s5.head()
```

```python
s5champs = s5["champion"].unique()[0:20]
```

```python
s5["pick_ban_ratio"][0:20]
```

```python
plt.bar(s5champs, s5["pick_ban_ratio"][0:20])
plt.xticks(s5champs, rotation = "vertical")
plt.xlabel("Champion")
plt.ylabel("Pick Ban Ratio (%)")
plt.title("Season 5")
plt.grid()
plt.show()
```

We can repeat this with the other seasons. Let's create a function that will return this graph. We can use our dictionary to do this.

```python
season_dictionary[5]["champion"].unique()[0:20]
```

```python
def find_op_champ(season):
    fig = plt.figure()
    plt.bar(season_dictionary[season]["champion"].unique()[0:20],
            season_dictionary[season]["pick_ban_ratio"][0:20])
    plt.xticks(season_dictionary[season]["champion"].unique()[0:20],
               rotation = "vertical")
    plt.xlabel("Champion")
    plt.ylabel("Pick Ban Ratio (%)")
    plt.title("Season " + str(season))
    plt.grid()
    # BE VERY CAREFUL WHEN RUNNING THIS; THIS WHILE SAVE FILES TO YOUR LOCAL FOLDER; UNCOMMENT IT IF YOU WANT A PNG OF THE GRAPHS
    # plt.savefig("find_op_champ_s" + str(season) + ".png", dpi = 150, bbox_inches="tight", facecolor='white', transparent=False)
    plt.show()
```

```python
# BE VERY CAREFUL WHEN RUNNING THIS; THIS WHILE SAVE FILES TO YOUR LOCAL FOLDER
find_op_champ(5)
find_op_champ(6)
find_op_champ(7)
find_op_champ(8)
find_op_champ(9)
find_op_champ(10)
find_op_champ(11)
find_op_champ(12)
```

### How do I find the most overpowered champions throughout all the seasons?

```python
champs.groupby(["season", "champion"]).sum()
```

Let's say I want to find only the champs with over a pick ban ratio of 90 for each season.

```python
over90 = champs.groupby(["season", "champion"]).sum()[champs.groupby(["season", "champion"]).sum()["pick_ban_ratio"] > 90]
over90
```

First, let's get the champion names and seasons. Use .index.get_level_values()

```python
champ_list = champs.groupby(["season", "champion"]).sum()[champs.groupby(["season", "champion"]).sum()["pick_ban_ratio"] > 90].index.get_level_values("champion")
champ_list
```

```python
season_list = champs.groupby(["season", "champion"]).sum()[champs.groupby(["season", "champion"]).sum()["pick_ban_ratio"] > 90].index.get_level_values("season")
season_list
```

```python
# reset from groupby so we can graph it by color
over90.reset_index(inplace = True)
over90
```

```python
groups = over90.groupby("season")

for season, group in groups:
    plt.scatter(group.champion, group["pick_ban_ratio"])

plt.xticks(champ_list, rotation = "vertical")
plt.xlabel("Champion")
plt.ylabel("Pick Ban Ratio (%)")
plt.title("Top 25 Pick Ban Ratios By Champion Per Season (S5-12)")
plt.grid()
plt.legend(labels = season_list.unique(), loc='upper right', bbox_to_anchor=(1.19, 1.025))
# plt.savefig will save to local file; uncomment it if you want this graph to be saved
# plt.savefig("highest_pbr_per_season.png", dpi = 150, bbox_inches="tight", facecolor='white', transparent=False)
plt.show()
```

### How do I find the champions that have consistently stayed at the top during worlds time?

```python
top_champs_ever = champs.groupby(["champion"]).sum().sort_values("games_contests", ascending = False)[0:25]
top_champs_ever
```

```python
top_champs_ever_list = champs.groupby(["champion"]).sum().sort_values("games_contests", ascending = False)[0:25].index.get_level_values("champion")
top_champs_ever_list
```

```python
plt.scatter(top_champs_ever_list, top_champs_ever["games_contests"])
plt.xticks(top_champs_ever_list, rotation = "vertical")
plt.xlabel("Champion")
plt.ylabel("Games Contested")
plt.title("Most Contested Champions in Worlds Seasons 5-12")
plt.grid()
# plt.savefig will save to local file; uncomment it if you want this graph to be saved
# plt.savefig("most_contested_champ.png", dpi = 150, bbox_inches="tight", facecolor='white', transparent=False)
plt.show()
```

### How do I find a champion's statistics throughout all of worlds?

```python
champs.groupby(["champion"]).sum().reset_index()
```

```python
champ_stats = champs.groupby(["champion"]).sum().reset_index()
champ_stats
```

```python
champ_stats["champion"]
```

```python
def find_champ(champ):
    return champ_stats[champ_stats["champion"] == champ]
```

```python
find_champ("Lee Sin")
```

### When worlds comes around, Riot usually asks us to predict what champions will have the highest/lowest statistics. What champion has the highest number of kills? Deaths? Creep score? Damage?

```python
champs.columns
```

### Highest Kills

```python
highest_kills = champs.groupby(["champion"]).sum().sort_values("kills", ascending = False)[0:25]
highest_kills
```

```python
champ_list_highest_kills = highest_kills.index.get_level_values("champion")
champ_list_highest_kills
```

```python
plt.scatter(champ_list_highest_kills, highest_kills["kills"])
plt.xticks(champ_list_highest_kills, rotation = "vertical")
plt.xlabel("Champion")
plt.ylabel("Kill Amount")
plt.title("Highest Kills by Champion in Worlds Seasons 5-12")
plt.grid()
# plt.savefig will save to local file; uncomment it if you want this graph to be saved
# plt.savefig("highest_kills.png", dpi = 150, bbox_inches="tight", facecolor='white', transparent=False)
plt.show()
```

### Highest Deaths

```python
highest_deaths = champs.groupby(["champion"]).sum().sort_values("deaths", ascending = False)[0:25]
highest_deaths
```

```python
champ_list_highest_deaths = highest_deaths.index.get_level_values("champion")
champ_list_highest_deaths
```
